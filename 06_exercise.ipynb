# **EXERCISE 6**

Create a basic level neural network and try to predict a value. Try to improve the model also.

```python
# Importing tensorflow
import tensorflow as tf
print("Tensorflow version we're using: ", tf.__version__)
```

```python
# Importing other dependencies
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Creating features
X = np.array([2, 5, 7, 10, 12, 15, 17, 20])

# Creating label
y = np.array([22, 25, 27, 30, 32, 35, 37, 40])

X, y
```

```python
# Converting NumPy arrays into Tensorflow tensors
X = tf.constant(X)
y = tf.constant(y)
X, y
```

```python
# For better precision let's convert int64 tensors into float32
X = tf.cast(X, dtype=tf.float32)
y = tf.cast(y, dtype=tf.float32)
X, y
```

```python
# Visualizing our data points
plt.figure(figsize=(15, 10))
plt.scatter(X, y)
plt.xlabel("X", fontsize=30)
plt.ylabel("y", fontsize=30)
plt.title("Visualizing X and y", fontsize=45)
```

## **Creating Neural Network**

Steps we follow are:
1. Creating a model
2. Compiling the model
3. Fitting the model

```python
# Let's create our first neural network

# Setting random seed for reproducability
tf.random.set_seed(42)

# 1. Creating a model
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(1),
])

# 2. Compiling the model
model.compile(loss = tf.keras.losses.MeanAbsoluteError(),
              optimizer = tf.keras.optimizers.SGD(),
              metrics=["mae"])

# 3. Fitting the model
model.fit(tf.expand_dims(X, axis=-1), y, epochs=5)  # epochs is basically laps
```

```python
# Checking for X and y
X, y
```

```python
# Let's predict the value
# The above pattern goes like 2+3+2+3+2......... , so when we are giving X=22, our predicted y should be 42

model.predict(tf.constant([22.0]))
```

## Improving our model

We have to imporve our model , as we are getting a output which we didn't expected.
Our output should be 42 or near about 42.

```python
# Let's create our model again

tf.random.set_seed(42)

# 1. Creating a model
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(1),
])

# 2. Compiling the model
model.compile(loss = tf.keras.losses.MeanAbsoluteError(),
              optimizer = tf.keras.optimizers.SGD(),
              metrics=["mae"])

# 3. Fitting the model
model.fit(tf.expand_dims(X, axis=-1), y, epochs=100)  # Increasing epochs for improvement
```

```python
# Let's check X and y
X, y
```

```python
# Now lets' try to predict
model.predict(tf.constant([22.0]))
```

Woah we got a nearbout number of 42.
Let's again try to improve our model more

### Improving model again

```python
# Let's again try to improve the model more

tf.random.set_seed(42)

# 1. Creating the model
model = tf.keras.models.Sequential([
    # Let's increase number of layers this time
    tf.keras.layers.Dense(1),
    tf.keras.layers.Dense(1),
    tf.keras.layers.Dense(1),
])

# 2. Compiling the model
model.compile(loss = tf.keras.losses.MeanAbsoluteError(),
              optimizer = tf.keras.optimizers.SGD(),
              metrics=["mae"])

# 3. Fitting the model
model.fit(tf.expand_dims(X, axis=-1), y, epochs=100)
```

```python
X, y
```

```python
# Let's try to predict now
model.predict(tf.constant([22.0]))
```

##### Woah we moved closed to 42
